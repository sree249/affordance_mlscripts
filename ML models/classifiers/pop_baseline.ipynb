```python
import pandas as pd
import numpy as np
```

```python
#device list by number of users 
device_list = ['Smart_Phone','Tablet','Laptop','Smart_Watch','Desktop_Computer','Smart_Speaker']
```

```python
counter = 0
tmp_list = []

#join tables with outer join and fill the empty columns with na

for i in device_list:
    
    tmp_list.append(i)
    
    if i =='Smart_Phone':
        big_table = pd.read_csv('predictions_'+str(i)+'.csv')[['ResponseId','actual_use_'+i]]
        print(i)
        print(big_table['actual_use_'+i].sum())
        big_table['Pop_Prediction_'+i] = 1
    else:
        tmp_table = pd.read_csv('predictions_'+str(i)+'.csv')[['ResponseId','actual_use_'+i]]
        print(i)
        print(tmp_table['actual_use_'+i].sum())
        big_table = pd.merge(tmp_table,big_table,on='ResponseId',how='outer')
        big_table['Pop_Prediction_'+i] = 1

for i in device_list:
    
    big_table['Pop_Prediction_'+i] = 1
        
```

```python
# total column is the number of device of each user in total
big_table['total'] = 6 - big_table.isna().sum(axis=1)
```

```python
big_table.head(55)
```

```python
#find the average popularity across all device
pop_percentage = np.mean(big_table['total']/6)
```

```python
pop_percentage
```

```python
# the number of devices to check for the participant's total number of devices 
big_table['tocheck'] = (big_table['total']*pop_percentage).apply(int)+1
```

```python
big_table.head(5)
```

```python
error_list = []
error_total = 0
True_Positive = 0
False_Positive = 0
Positive = 0
False_Negative = 0
test = 0
for index, row in big_table.iterrows():
    
    count = row['tocheck']
    
    
    error = 0
    
    for i in range(0,6):
        
        for j in device_list:
            
            if count == 0:
                break
            
            Positive += 1
            
            #if the user own the device (is null) then we check this device
            
            if not pd.isnull(row['actual_use_'+j]): 
                
                if row['actual_use_'+j] == 0:
                    error +=1
                else:
                    True_Positive +=1
                    
                count = count -1
            
    count = 0
    
    #caculate false negative
    
    for i in range(0,6):
        
        for j in device_list:
            if count < row['tocheck']:
                count = count +1
                continue
            
            #if the user own the device (is null) then we check this device and this device is not "popular" for the participant 
            
            if not pd.isnull(row['actual_use_'+j]):
                
                test +=1
                
                if row['actual_use_'+j] == 1:
                    False_Negative +=1
                count = count +1

    error_total += error/row['tocheck']
    error_list.append(error/row['tocheck'])
                
    
```
